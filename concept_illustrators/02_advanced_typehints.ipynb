```python
from typing import List, Tuple, Dict
```

```python
# Before python 3.10 we used to do like below

price: List[int] = [1,2,3]
price: Tuple[int,int,int] = (4,5,6)
price: Dict[str, int] = {
    "item1" :43,
    "item2" : 78
}
```

```python
price: List[int] = [1,2,3]
print(price)
```

```python
price: Dict[str, int] = {
    "item1" :43,
    "item2" : 78
}

print(price)
```

- Now with Python 3.10 onwards we can directly use list, tuple, dict keywords from python itself, no need to import from typing module

```python

price: list[int] = [1,2,3,5]
print(price)
```

```python
price: tuple[int,int, float] = (3, 5, 9.8)
print(price)
```

```python
price: dict[str, int] = {
    "item1" : 34, 
    "item2" : 65
}
print(price)
```

- Common scenario where we want to have a list with integers and some floating point numbers

```python
from typing import Union, List

x: List[Union[int, float]] = [3.5, 6, 9.8]
print(x)
```

OR with Python 3.10 we can also put it as like below

```python
x: List[int | float ] = [3.5, 6, 9.8]
print(x)
```
