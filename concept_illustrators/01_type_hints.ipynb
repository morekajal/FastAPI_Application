```python
# If we dont use any type hinting

def total_price(price1, price2):
    print(f'Price 1: {price1}, Price 2: {price2}, Sum: {price1 + price2}')
    return f'Total Price: {price1+price2}'
```

```python
sum_x = total_price(10, 25)
print(sum_x)
```

We can see in the above function, it is by default taking input as string, what if i want to add with some float

```python
sum_x = total_price(101, 4.56)
print(sum_x)
```

But what if we want to add strings

```python
sum_x = total_price('10', 25)
print(sum_x)
```

```python
sum_x = total_price('10', '25')
print(sum_x)
```

But now, we can type hint this function deciding what kind of input it should take.

```python
def total_price(price1, price2) -> str:
    print(f'Price 1: {price1}, Price 2: {price2}, Sum: {price1 + price2}')
    return f'Total Price : {price1 + price2}'
```

```python
sum_y = total_price('10', '25')
```

```python
def total_price(price1: float, price2: float) -> str:
    # the defined means - it expects floating point value input and returns a string type
    print(f'Price 1: {price1}, Price 2: {price2}, Sum: {price1 + price2}')
    return f'Total Price : {price1 + price2}'

sum_z = total_price(10.34, 55.6)
```
